```python
# 문제1
students = {"민수": [80, 70, 90], "영희": [40, 55, 60], "지수": [100, 95, 90], "철수": [30, 45, 20]}


# 1. 한 학생의 평균 점수를 계산하는 함수
def calculate_average(scores):
    # scores 리스트의 모든 점수를 더한 후 개수로 나눈다
    return sum(scores) / len(scores)


# 1번 함수 테스트
# students 딕셔너리에서 "민수"의 점수 리스트를 가져와서 평균 계산
minsu_average = calculate_average(students["민수"])
print(f"민수의 평균: {minsu_average}")
print()  # 빈 줄 출력, 섹션 구분


# 딕셔너리에 값을 추가/저장하는 기본 문법: 딕셔너리이름[key] = value (함수 실행 결과)
# 2. 모든 학생의 평균을 딕셔너리로 만드는 함수
def calculate_averages(students_dict):
    # 빈 딕셔너리를 만든다
    averages = {}
    # students_dict를 순회하면서 이름(key)과 점수 리스트(value)를 가져온다
    for name, scores in students_dict.items():
        # key는 학생 이름, value는 평균 점수를 저장한다
        # 대괄호 []: 딕셔너리의 key를 지정할 때 사용
        averages[name] = calculate_average(scores)  # 함수 호출: calculate_average(scores)는 평균값을 계산해서 반환
    # 완성된 평균 딕셔너리를 반환한다
    return averages


# 3. 평균이 60 이상인 학생만 새 딕셔너리로 반환하는 함수
def filter_passing_students(students_dict):
    # 빈 딕셔너리를 만든다 (합격자를 저장할 딕셔너리)
    passing_students = {}
    # students_dict를 순회하면서 이름(key)과 점수 리스트(value)를 가져온다
    for name, scores in students_dict.items():
        # 현재 학생의 평균 점수를 계산한다
        average = calculate_average(scores)
        # 평균이 60 이상이면
        if average >= 60:
            # key는 학생 이름, value는 평균 점수를 저장한다
            passing_students[name] = average
    # 합격자 딕셔너리를 반환한다
    return passing_students


# 4. 평균이 가장 높은 학생의 이름을 반환하는 함수
def get_top_student(students_dict):
    # 최고 평균 학생의 이름을 저장할 변수 (초기값 None)
    top_student = None
    # 최고 평균 점수를 저장할 변수 (초기값 0)
    highest_average = 0
    # students_dict를 순회하면서 이름(key)과 점수 리스트(value)를 가져온다
    for name, scores in students_dict.items():
        # 현재 학생의 평균 점수를 계산한다
        average = calculate_average(scores)
        # 현재 평균이 지금까지의 최고 평균보다 높으면
        if average > highest_average:
            # 최고 평균을 현재 평균으로 갱신한다
            highest_average = average
            # 최고 평균 학생을 현재 학생으로 갱신한다
            top_student = name
    # 최고 평균 학생의 이름을 반환한다
    return top_student


# 최종 결과 출력
# 2번 함수를 호출하여 모든 학생의 평균을 계산한다
averages = calculate_averages(students)
print("모든 학생의 평균:", averages)

# 3번 함수를 호출하여 평균 60 이상인 학생만 필터링한다
passing_students = filter_passing_students(students)
print("평균 60 이상인 학생:", passing_students)

# 4번 함수를 호출하여 최고 평균 학생을 찾는다
top_student = get_top_student(students)
print("최고 평균 학생:", top_student)
```

```python
# 은행 계좌 시뮬레이터

# 계좌 데이터
accounts = {"A001": 50000, "A002": 120000, "A003": 30000}


# 1. 입금 함수 : : 계좌번호 찾아서 금액 더하기(딕셔너리[키]: 값을 가져오거나 수정)
# 입금하려면 뭐가 필요할까?
# 1. 어떤 계좌 목록(딕셔너리)
# 2. 어느 계좌 번호
# 3. 얼마를 입금할지
def deposit(accounts_dict, account_id, amount):  # 함수정의(뭘 받을지): 대상 → 위치 → 값
    # accounts_dict: 전체 계좌 정보가 담긴 딕셔너리
    # account_id: 입금할 계좌 번호 (예: "A001")
    # amount: 입금할 금액
    accounts_dict[account_id] = accounts_dict[account_id] + amount  # 함수본문(실제로 할 일)
    # 딕셔너리[키] = 딕셔너리[키] + 입금액: 현재 잔액에 입금액을 더해서 다시 저장
    # (통장) = (먼저 계좌에 얼마가 있는지 계산) + (입금할 금액)
    # (저장할 위치) = (계산할 값)
    # accounts["A001"]: ~라는 key를 찾아가서 그 key의 value를 가져오거나 수정하는 것


# 2. 출금 함수 (잔액 부족 시 출금 실패 처리): 잔액 확인 → 충분하면 빼기, 부족하면 False 반환
# 출금하려면 뭐가 필요할까?
# 1. 어떤 계좌 목록(딕셔너리)
# 2. 어느 계좌 번호
# 3. 얼마를 출금할지
# 4. 잔액이 부족하면? → 실패 처리 필요
def withdraw(accounts_dict, account_id, amount):
    # 출금 전에 먼저 잔액이 충분한지 확인해야 함
    if accounts_dict[account_id] >= amount:  # 잔액 >= 출금액이면
        accounts_dict[account_id] = accounts_dict[account_id] - amount
        # 현재 잔액에서 출금액을 빼서 다시 저장
        return True  # 출금 성공했다는 신호
    else:  # 잔액이 부족하면
        return False  # 출금 실패했다는 신호


# 3. 전체 잔액 합계 계산 함수 :  .values()로 모든 잔액 가져와서 하나씩 더하기(.values(): 딕셔너리의 값들만)
# 모든 계좌의 잔액을 다 더하면 됨
def total_balance(accounts_dict):
    total = 0  # 합계를 저장할 변수 (처음엔 0)

    # 딕셔너리의 모든 값(value)들을 하나씩 꺼내서 더하기
    for balance in accounts_dict.values():  # .values()는 딕셔너리의 값들만 가져옴
        total = total + balance  # 현재까지 합계 + 이번 잔액

    return total  # 최종 합계를 반환


# 4. 잔액이 50000 미만인 계좌만 모아 반환하는 함수: .items()로 (계좌번호, 잔액) 꺼내서 50000 미만이면 리스트에 추가
# .items(): 딕셔너리의 (키, 값) 쌍으로
# 위험 계좌(잔액이 적은 계좌)를 찾아서 리스트로 반환
def find_low_balance(accounts_dict, threshold=50000):
    # threshold: 기준 금액 (기본값 50000)

    low_accounts = []  # 위험 계좌를 담을 빈 리스트

    # 딕셔너리의 모든 항목(계좌번호, 잔액)을 하나씩 확인
    for account_id, balance in accounts_dict.items():  # .items()는 (키, 값) 쌍으로 가져옴
        if balance < threshold:  # 만약 잔액이 기준금액보다 적으면
            low_accounts.append(account_id)  # 그 계좌번호를 리스트에 추가

    return low_accounts  # 위험 계좌 리스트 반환


# 실행
deposit(accounts, "A001", 20000)  # A001에 20000원 입금
print(f"A001 잔액: {accounts['A001']}")  # 입금 후 잔액 확인

result = withdraw(accounts, "A003", 50000)  # A003에서 50000원 출금 시도
if not result:  # result가 False면 (출금 실패하면)
    print("A003 출금 실패")

print(f"전체 은행 자산: {total_balance(accounts)}")  # 모든 계좌 합계
print(f"위험 계좌: {find_low_balance(accounts)}")  # 잔액 50000 미만 계좌들
```
