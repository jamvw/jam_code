````python
class Person:
    def __init__(self, name, age):  # 생성자
        self.name = str(name)  # 인스턴스 변수 #클래스 자신 안의 함수를 쓰겠다: self.
        self.age = str(age)  # 인스턴스 변수
        ## return 새로만든 붕어빵의 주솟값

    def introduce(self):
        print(f"안녕하세요, 제 이름은 {self.name}이고, 나이는 {self.age}세입니다.")


person1 = Person("Alice", 25)  # person1 = 새로찍어낸 붕어빵의 주솟값
# 붕어빵 = 붕어빵틀(로 찍어냄)

print(person1.name)

person1.introduce()
````

- **인스턴스 변수(instance variable)**
    - 각 객체마다 별도로 관리되는 변수

    - ex) p1의 name은 Alice, p2의 name은 Bob

    **붕어빵마다의 고유변수**

- **클래스 변수(class variable)**
    - 클래스로부터 만들어진 모든 인스턴스가 공유하는 변수.
    - 클래스 블록 내에서 `self` 없이 바로 변수를 정의하면 클래스 변수로 설정됩니다.
    - ex) 생성자에 self.type이 아니라 type=’human’으로 정의되었다 → p1의 type은 ‘human’, p2의 type도 ‘human’

    **공용 변수**

````python
class Car:
    wheels = 4  # 클래스 변수: 모든 Car 인스턴스는 바퀴가 4개

    def __init__(self, color):
        self.color = color  # 인스턴스 변수: 각 차마다 색이 다를 수 있음


car1 = Car("red")
car2 = Car("blue")

print(car1.color)  # "red" (car1만의 인스턴스 변수)
print(car2.color)  # "blue" (car2만의 인스턴스 변수)

print(car1.wheels)  # 4 (클래스 변수는 모든 객체가 공유)
print(car2.wheels)  # 4
````

- 여기서 wheels는 클래스 변수로 Car의 모든 인스턴스에 동일한 값(4개)을 갖습니다.

## 5. 메서드의 종류

1. **인스턴스 메서드(Instance Method)**
    
    → 붕어빵 고유의 변수(인스턴스 변수)를 다루는 함수!
    
    - 첫 번째 매개변수로 `self`를 받습니다.
    - 각 인스턴스에 대해서 동작하며, 인스턴스 변수를 다루는 데 주로 사용됩니다.
    
    '''

    class Person:
        def __init__(self, name):
            self.name = name
    
        def say_hello(self): #인스턴스 메서드
            print(f"Hello, my name is {self.name}")
    
    person1 = Person("alice")
    person1.say_hello()
    
    '''
    
2. **클래스 메서드(Class Method)**
    
    → 붕어빵 틀자체(클래스)에 대한 것을 다루는 함수!
    
    - 첫 번째 매개변수로 `cls`를 받습니다. (`self` 대신 `cls` 키워드 사용)
    - 데코레이터 `@classmethod`를 사용합니다.
    - 클래스 자체를 인자로 받습니다. (클래스=틀 을 다뤄야하니까)
    - 활용: 클래스 변수를 다루는 상황 / 새로운 인스턴스를 생성하는 메서드 등을 정의
    
    '''

    class Person:
        count = 0
    
        def __init__(self, name):
            self.name = name
            Person.count += 1
    
        @classmethod
        def how_many(cls):
            print(f"지금까지 {cls.count}명이 만들어졌습니다.")
    
    p1 = Person("Alice")
    p2 = Person("Bob")
    Person.how_many()  # "지금까지 2명이 만들어졌습니다."
    
    '''
    
3. **정적 메서드(Static Method)**
    
    → 걍 함수!
    
    - 첫 번째 매개변수로 `self`나 `cls`를 받지 않습니다.
    - 사실 클래스에 굳이 소속될 필요가 없다. 그런데 “분류상” 틀에 넣어야 깔끔해서 넣는 경우. (주로 클래스나 인스턴스 변수에 접근할 필요가 없는 경우)
    - 데코레이터 `@staticmethod`를 사용합니다.
    
    '''

    class MathUtils:
        @staticmethod
        def add(a, b):
            return a + b
    
    result = MathUtils.add(3, 5)
    print(result)  # 8
    
    '''

````python
# Python 클래스의 메서드에서 return의 필요성은 메서드의 목적에 따라 다릅니다.
# Return이 필요한 경우
class Calculator:
    def add(self, a, b):
        return a + b  # 결과값을 반환해야 사용 가능

    def get_name(self):
        return "Calculator"


calc = Calculator()
result = calc.add(5, 3)  # return이 있어야 result에 값이 담김
print(result)  # 8


# Return이 필요 없는 경우
class Person:
    def __init__(self, name):
        self.name = name
        # return 없음 - 인스턴스 변수만 설정

    def set_age(self, age):
        self.age = age
        # return 없음 - 상태만 변경


person = Person("철수")
person.set_age(25)
print(person.age)  # 25 - self를 통해 접근 가능

# 핵심 차이점

# self를 통한 저장: 인스턴스 변수(self.name 등)에 저장하면 나중에 객체.변수로 접근 가능 → return 불필요
# 값을 즉시 사용: 메서드 호출 결과를 바로 사용하려면 return 필요


class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1
        # return 없음 - self.count에 저장됨

    def get_count(self):
        return self.count
        # return 필요 - 값을 가져와야 함


c = Counter()
c.increment()
print(c.count)  # 1 (self로 접근)
print(c.get_count())  # 1 (return으로 받음)

# 자동으로 주소를 찾아주는 건 맞지만, 메서드가 계산한 결과를 즉시 사용하려면 return이 필요!
````

````python
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def get_average(self):
        if not self.scores:
            return 0
        return sum(self.scores) / len(self.scores)

    def introduce(self):
        print(f"이름: {self.name}, 학번: {self.student_id}, 평균점수: {self.get_average()}")


# 사용 예시
s1 = Student("Alice", "20230001")
s1.add_score(90)
s1.add_score(85)
s1.add_score(100)

s2 = Student("Bob", "20230002")
s2.add_score(70)
s2.add_score(75)

s1.introduce()  # 이름: Alice, 학번: 20230001, 평균점수: 91.666...
s2.introduce()  # 이름: Bob, 학번: 20230002, 평균점수: 72.5
````

## 정리

- 클래스는 객체(인스턴스)를 만들기 위한 설계도이며, OOP의 핵심 개념입니다.
- 인스턴스 변수, 클래스 변수를 통해 데이터를 관리하고, 인스턴스 메서드, 클래스 메서드, 정적 메서드를 통해 기능을 구현합니다.

- 클래스를 통해 관련된 데이터와 기능을 논리적 단위로 묶어내면 큰 규모의 프로그램을 더 체계적이고 유지 관리하기 쉽게 만들 수 있습니다.

    1. `move(dx, dy)` → 현재 좌표를 `dx`, `dy`만큼 이동시킴
    2. `show()` → 현재 좌표를 `"Point(x, y)"` 형식으로 출력

**예시 실행**

```python
p = Point(1, 2)
p.move(2, 3)     # (1+2, 2+3) → (3, 5)
p.show()         # Point(3, 5)
```

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def move(self, dx, dy):
        self.x += dx
        self.y += dy

    def show(self):  # self를 넣어주는 이유
        print(f"Point{self.x},{self.y}")


p = Point(1, 2)
p.move(2, 3)  # (1+2, 2+3) → (3,5)
p.show()  # Point(3,5)
````

- 객체가 생성될 때마다 `count`를 1 증가시키세요.
- 클래스 메서드 `how_many()`를 만들어 현재까지 몇 개의 인스턴스가 만들어졌는지 출력하세요.

**예시 실행**

```python
a = Counter()
b = Counter()
c = Counter()
Counter.how_many()   # 현재까지 3개 생성
```

````python
class Counter:
    count = 0 #클래스 변수
    
    def __init__self:
      Counter.count += 1
    
    @classmethod
    def how_many(cls):
        print(f"현재까지 만들어진 인스턴스는 {clas.count}개입니다")
        
a = Counter()
b = Counter()
c = Counter()

Counter.how_many()  #현재까지 3개 생성
````

## 문제 3 | Temperature 클래스

**문제 설명**

- `Temperature`라는 이름의 클래스를 정의하세요.
- 인스턴스 변수 `celsius`에 섭씨 온도를 저장합니다.
- 메서드 기능은 아래와 같습니다.
    1. `to_fahrenheit()` → 섭씨를 화씨로 변환하여 반환 (`℉ = ℃ × 9/5 + 32`)
    2. `is_boiling()` → 섭씨 100도 이상이면 `True`, 아니면 `False` 반환

**예시 실행**

```python
t = Temperature(36.5)
print(t.to_fahrenheit())  # 97.7
print(t.is_boiling())     # False

t2 = Temperature(120)
print(t2.is_boiling())    # True
```

````python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius
    
    def to_fahrenheit(self):
        return self.celsius * 9/5 + 32
    
    def is_boiling(self):
        if self,celsius >= 100 : return True
        else: return False

t = Temperature(36.5)
print(t.to_fahrenheit())  # 97.7
print(t.is_boiling())

t2 = Temperature(120)
print(t2.is_boiling())
````
