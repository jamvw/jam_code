```python
# 문제1
def bigger(a, b):
    # a와 b 중 어느 것이 더 큰지 비교
    if a > b:  # a가 b보다 크면
        return a  # a를 반환
    else:  # a가 b보다 크지 않으면 (a <= b인 경우)
        return b  # b를 반환


# 함수 호출 방법 비교

# 방법 1: 반환값을 저장하지 않고 그냥 호출
# bigger(3, 5)
# → 함수는 실행되고 5를 반환하지만, 그 값을 어디에도 저장하지 않음
# → 반환값이 사라져서 사용할 수 없음

# 방법 2: 반환값을 변수에 저장
res = bigger(3, 5)  # 함수가 반환한 5를 res 변수에 저장
print(res)  # 저장된 값을 출력 → 5가 출력됨
```

```python
# 문제4
def double(n):
    return n * 2  # n*2라는 연산만 하고 함수값은 반환되지 않았기 떄문에 return을 사용


result = double(5)
print(result)

# return은 값을 돌려주기만 함
# 돌려받은 값을 변수에 저장해야 나중에 사용 가능
# 저장 안 하면 반환값은 사라짐 (휘발됨)
```

```python
# 문제5
# 리스트, 딕셔너리, 튜플 = 함수 안에서도 수정 가능
# int, float, Bool = 함수 안에서 수정이 불가능


def change_first(x):
    x[0] = 0  # 리스트는 가변객체이므로 내부를 수정할 수 있음


nums = [1, 2, 3]
change_first(nums)
print(nums)

# 실행결과에서 원래 리스트가 바뀌지 않는 이유는 함수 내부에서 x = [0, 2, 3]을 실행하면
# x라는 지역 변수가 리스트 [0, 2, 3]를 가리키게 되는데,
# 그렇게 되면 원래 nums가 가리키던 리스트 [1, 2, 3]와의 연결이 끊어지지만,
# 함수가 끝나면 지역 변수 x는 사라지고, print(nums)를 실행하게 됩니다.
# print(nums)이 실행되는 시점에는 원래 리스트가 바뀌지 않은 상태입니다.
# 따라서 출력 결과는 nums 리스트의 값인 [1,2,3]이 됩니다.
```

```python
# 문제6
# 리스트, 딕셔너리, 튜플 = 함수에 주소값으로 전달


# 정수(int)는 불변(immutable) 객체이기 때문에 함수 내부에서 n=n+1을 실행하면
# 새로운 정수 객체가 생성되고, 지역변수 n만 이를 가리키게 됩니다.
# 원본 변수 x는 여전히 10을 가리키므로 함수 실행 후에도 x는 변하지 않습니다
# 반환값을 변수 x 에 할당해줘야 x의 변경된 값을 사용할 수 있습니다.
def add_one(n):
    # n = n + 1
    return n + 1


x = 10
x = add_one(x)  # 반환값을 x에 할당하기
print(x)  # 출력: 11
```

```python
# 문제7
# 리스트는 가변( mutable) 객체이기 때문에 함수 내부에서 lst = [0, 0, 0]을 실행하면
# lst라는 지역 변수가 새로운 리스트 [0, 0, 0]을 가리키게 됩니다.
# 이렇게 되면 원래 nums가 가리키던 리스트와의 연결이 끊어지지만,
# 함수가 끝나면 지역 변수 lst는 사라지고, print(nums)를 실행하게 됩니다.
# print(nums)이 실행되는 시점에는 원래 리스트가 바뀌지 않은 상태입니다.
# 따라서 출력 결과는 nums 리스트의 값인 [1, 2, 3]이 됩니다.
# reset 함수 내부에서 lst 리스트의 내용을 직접 수정하려면
# lst[0] = 0, lst[1] = 0, lst[2] = 0 과 같이 각 요소를 개별적으로 변경해야 합니다.


def reset(lst):
    lst[0] = 0
    lst[1] = 0
    lst[2] = 0


nums = [1, 2, 3]
reset(nums)
print(nums)
```

```python
# 문제8
# 왜 문자열이 바뀌지 않는가?
# 문자열은 불변(immutable) 객체이기 때문에
# 함수 내부에서 s.upper()를 실행하면 새로운 문자열 객체가 생성되고,
# 지역변수 s만 이를 가리키게 됩니다.
# 원본 변수 text는 여전히 "hello"를 가리키므로 함수 실행 후에도 text는 변하지 않습니다.
# 따라서 출력 결과는 원래 문자열 "hello"가 됩니다.


def make_upper(s):
    s.upper()


text = "hello"
make_upper(text)
print(text)
```

```python
# 문제9
import math


def calculate_circle_area(radius):
    area = math.pi * radius**2
    return area


radius = 3
area = calculate_circle_area(radius)
print("반지름이" + str(radius) + "인 원의 넓이:", area)
```
