```python
# 1. 매개변수 X, return X : 그냥 실행만 하는 함수 - 저장 안 함, 바로 출력
# 함수 내부에 출력 명령(print)을 작성해서, 함수를 호출할 때 그 출력값이 바로 나오도록 한다.
def greet():
    print("안녕하세요!")


greet()  # 안녕하세요!


# 2. 매개변수 O, return X : 입력받아서 실행하는 함수 - 받은 값으로 바로 출력
# 함수에 매개변수를 추가하고, 함수 내부에서 매개변수를 포함한 내용을 출력하도록 작성한다.
# 함수를 호출할 때 인자를 전달하면, 그 값이 포함된 결과가 화면에 바로 출력된다.
def greet(name):
    print(f"{name}님, 안녕하세요!")


greet("철수")  # 철수님, 안녕하세요!


# 3. 매개변수 X, return O : 결과를 돌려주는 함수 - 값을 변수에 담아서 사용
# 함수에 내부에 반환값을 작성하고, 변수에 함수의 실행 결과를 할당한 다음,
# 그 변수를 출력했을 때 함수의 반환값(문자열)이 나오도록 한다.
def get_greeting():
    return "안녕하세요!"


message = get_greeting()
print(message)  # 안녕하세요!


# 4. 매개변수 O, return O : 인자들을 입력받아서 결과를 돌려주는 함수 - 계산 후 값 반환
# 여러 개의 매개변수(인자)들의 연산 결과를 반환하도록 작성한다
# 변수에 함수 실행 결과를 할당해서, 그 변수를 출력했을 때 계산된 반환값이 나오도록 한다.
def add(a, b):
    return a + b


result = add(5, 3)
print(result)  # 8

# **실전 비교 예시**


# 1. 매개변수 X, return X - 단순 출력
def show_menu():
    print("1. 진라면")
    print("2. 신라면")


# 2. 매개변수 O, return X - 입력받아 출력만
def cook_ramen(ramen):
    print(f"{ramen} 조리 중...")


# 3. 매개변수 X, return O - 고정값 반환
def get_default_ramen():
    return "신라면"


# 4. 매개변수 O, return O - 입력받아 결과 반환
def prepare_ramen(ramen, spicy_level):
    return f"{ramen} (매운맛 {spicy_level}단계)"


# 사용

show_menu()  # 함수 호출해 메뉴 출력
cook_ramen("진라면")  # 진라면 조리 중...
# 함수 호출 시 "진라면"을 인자로 전달하면 매개변수 ramen에 할당됨
# → 매개변수에 '값'을 전달한다/'인자'로 전달한다

default = get_default_ramen()  # "신라면"
# "함수를 호출하고, 그 반환값을 변수에 저장한다"

final = prepare_ramen("진라면", 3)  # "진라면 (매운맛 3단계)"
# 두 개의 인자를 전달하여 함수를 호출하고, 반환값을 변수에 저장
```

매개변수가 필요한 이유:

1. 같은 로직으로 다른 결과를 만들기 위해
   - 라면 자판기는 하나지만, 어떤 라면을 선택하느냐에 따라 결과가 달라짐
   - 함수의 구조(로직)은 동일하지만, 입력값에 따라 다양한 결과를 생성할 수 있음

2. 함수의 재사용성을 높이기 위해
   - 매개변수 없이 고정된 값만 처리하면, 같은 작업을 여러 번 반복해야 함
   - 매개변수를 사용하면 하나의 함수로 여러 상황에 대응 가능

3. 함수를 범용적으로 만들기 위해
   - 특정 값에 의존하지 않고, 다양한 값을 받아서 처리할 수 있음
   - 코드의 유연성과 확장성이 높아짐

4. 입력값에 따라 동적으로 동작하게 하기 위해
   - 사용자 입력, 다른 함수의 결과 등 다양한 값을 받아 처리
   - 프로그램이 상황에 맞게 반응할 수 있게 함

```python
# 매개변수 없음 - 재사용성 낮음
def greet_john():
    print("John님, 안녕하세요!")


def greet_mary():
    print("Mary님, 안녕하세요!")


def greet_tom():
    print("Tom님, 안녕하세요!")


# 매개변수 있음 - 재사용성 높음
def greet(name):
    print(f"{name}님, 안녕하세요!")


greet("John")
greet("Mary")
greet("Tom")
```

함수 == 내가 원할 때 언제든지 호출해서 
같은 기능, 같은 로직을 가진 것을 
반복하는 코드

함수1 <- 라면 자판기

- 진라면 / 신라면 / 레드 / 안성탕면  <- 매개변수

- 어떤 값을 넣느냐에 따라서 결과값은 달라질 수 있지만,
로직(함수)은 같다.

함수2 <- 라면 자판기 밑에 볶음밥 기계

함수2에 함수1의 결과값을 쓰고싶을 때 함수1에서 return(반환)해준다.

```python
# 함수1 - 라면 자판기
def 라면자판기(라면종류):  # 매개변수
    if 라면종류 == "진라면":
        return "매운 진라면 완성!"
    elif 라면종류 == "신라면":
        return "얼큰한 신라면 완성!"
    else:
        return f"{라면종류} 완성!"


# 함수2 - 볶음밥 기계
def 볶음밥기계(재료):
    return f"{재료}를 넣은 볶음밥 완성!"


# 함수1의 결과를 함수2에 사용
라면 = 라면자판기("진라면")  # "매운 진라면 완성!" 반환
볶음밥 = 볶음밥기계(라면)  # 라면을 재료로 사용

print(볶음밥)  # 매운 진라면 완성!를 넣은 볶음밥 완성!
```
