```python
class Cat:
    hp = 100

    def __init__(self, name, age):
        self.name = (name,)
        self.age = (age,)

    def hi(self):
        print("야옹")

    def jump(self):
        print("휴먼 따라오거라")


cat1 = Cat("Momo", 5)  # 클래스를 호출해서 이름을 모모, 나이는 5살로 되어있는 애를 인스턴스화해서 객체로 만든다.
cat2 = Cat("Podo", 3)

cat1.hi()
cat2.hi()
cat1.jump()
cat2.jump()
```
