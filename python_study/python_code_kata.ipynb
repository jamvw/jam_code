```python
# 5
# 숫자 비교하기
# 0 ≤ num1 ≤ 10,000
# 0 ≤ num2 ≤ 10,000
# 정수 num1과 num2가 매개변수로 주어집니다. 두 수가 같으면 1 다르면 -1을 retrun하도록 solution 함수


def solution(num1, num2):
    return 1 if num1 == num2 else -1


num1, num2 = (9, 19)
print(solution(num1, num2))
```

```python
# 8
# 각도기
# 예각 : 0 < angle < 90
# 직각 : angle = 90
# 둔각 : 90 < angle < 180
# 평각 : angle = 180
# 0 < angle ≤ 180 / angle은 정수입니다.
# 각 angle이 매개변수로 주어질 때 예각일 때 1, 직각일 때 2, 둔각일 때 3, 평각일 때 4를 return하도록 solution 함수


def solution(angle):
    if 0 < angle < 90:
        return 1
    elif angle == 90:
        return 2
    elif 90 < angle < 180:
        return 3
    elif angle == 180:
        return 4


angle = 29
print(solution(angle))
```

```python
# 9
# 짝수의 합
# 0 < n ≤ 1000
# 정수 n이 주어질 때, n이하의 짝수를 모두 더한 값을 return 하도록 solution 함수


def solution(n):
    return sum(range(2, n + 1, 2))


n = 4
print(solution(n))
```

```python
# 10
# 0 ≤ numbers의 원소 ≤ 1,000
# 1 ≤ numbers의 길이 ≤ 100
# 정답의 소수 부분이 .0 또는 .5인 경우만 입력으로 주어집니다.
# 정수 배열 numbers가 매개변수로 주어집니다. numbers의 원소의 평균값을 return하도록 solution 함수


def solution(numbers):
    return sum(numbers) / len(numbers)


numbers = [89, 90, 91, 92, 93, 94, 95, 96, 97, 98, 99]
print(solution(numbers))
```

```python
# 11
# 정수 num이 짝수일 경우 "Even"을 반환하고 홀수인 경우 "Odd"를 반환하는 함수
# num은 int 범위의 정수
# 0은 짝수


def solution(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"


num = 4
print(solution(num))
```

```python
# 12
# 평균 구하기
# 정수를 담고 있는 배열 arr의 평균값을 return하는 함수
# arr은 길이 1 이상, 100 이하인 배열
# arr의 원소는 -10,000 이상 10,000 이하인 정수


def solution(arr):
    answer = sum(arr) / len(arr)
    return answer


arr = [1, 2, 3, 4]
print(solution(arr))
```

```python
# 13
# 자릿수 더하기
# 자연수 N이 주어지면, N의 각 자릿수의 합을 구해서 return
# N의 범위 : 100,000,000 이하의 자연수


def solution(N):
    s = str(N)
    N2 = list(s)
    result = sum(int(i) for i in N2)
    return result


N = 987
print(solution(N))
```

```python
# 14
# 약수의 합
# 정수 n을 입력받아 n의 약수를 모두 더한 값을 리턴하는 함수
# n은 0 이상 3000이하인 정수


def solution(n):
    result = []
    for i in range(1, n + 1):
        if n % i == 0:
            result.append(i)

    answer = sum(result)
    return answer


n = 12
print(solution(n))
```

```python
# 15
# 나머지가 1이 되는 수 찾기
# 자연수 n이 매개변수로 주어집니다. n을 x로 나눈 나머지가 1이 되도록 하는 가장 작은 자연수 x를 return
# 답이 항상 존재함은 증명될 수 있다
# 3 ≤ n ≤ 1,000,000


def solution(n):
    for i in range(1, n + 1):
        if n % i == 1:
            return i


n = 12
print(solution(n))
```

```python
# 16
# x만큼 간격이 있는 n개의 숫자
# 정수 x와 자연수 n을 입력 받아,x부터 시작해 x씩 증가하는 숫자를 n개 지니는 리스트를 리턴
# x는 -10000000 이상, 10000000 이하인 정수
# n은 1000 이하인 자연수


def solution(x, n):
    answer = []
    for i in range(1, n + 1):
        answer.append(x * i)
    return answer


x, n = (-4, 2)
print(solution(x, n))
```

```python
# 17
# 자연수 뒤집어 배열로 만들기
# 자연수 n을 뒤집어 각 자리 숫자를 원소로 가지는 배열 형태로 리턴
# n은 10,000,000,000이하인 자연수


def solution(n):
    answer = []

    # 1. n을 문자로 바꾸고("12345") -> 뒤집습니다("54321")
    reversed_str = str(n)[::-1]

    # 2. 뒤집힌 문자열에서 하나씩(i) 꺼냅니다. ('5', '4', '3'...)
    for i in reversed_str:
        # 3. 다시 숫자로 바꿔서 담습니다.
        answer.append(int(i))

    return answer


n = 12345
print(solution(n))
```

```python
# 18
# 문자열을 정수로 바꾸기
# 문자열 s를 숫자로 변환한 결과를 반환하는 함수
# s의 길이는 1 이상 5이하입니다.
# s의 맨앞에는 부호(+, -)가 올 수 있습니다.
# s는 부호와 숫자로만 이루어져있습니다.
# s는 "0"으로 시작하지 않습니다.


def solution(s):
    num = int(s)
    return num


s = "-1234"
print(solution(s))

# str은 부호(+,-)와 숫자로만 구성되어 있고, 잘못된 값이 입력되는 경우는 없습니다
```

```python
# 19
# 정수 제곱근 판별
# 임의의 양의 정수 n에 대해, n이 어떤 양의 정수 x의 제곱인지 아닌지 판단하려 합니다.
# n이 양의 정수 x의 제곱이라면 x+1의 제곱을 리턴, n이 양의 정수 x의 제곱이 아니라면 -1을 리턴하는 함수
# n은 1이상, 50000000000000 이하인 양의 정수


def solution(n):
    x = int(n**0.5)  # x는 n의 0.5제곱이다.
    if x**2 == n:
        return (x + 1) ** 2
    else:
        return -1


n = 121
print(solution(n))  # 144
```

핵심은 x에 n의 제곱근을 저장해야 한다는 점

```python
import math


def solution(n):
    x = int(math.sqrt(n))  # n의 제곱근을 정수로 변환
    if x**2 == n:
        return (x + 1) ** 2
    else:
        return -1


n = 121
print(solution(n))  # 144
```

```python
# 20
# 정수 내림차순으로 배치하기
# 함수 solution은 정수 n을 매개변수로 입력받습니다.
# n의 각 자릿수를 큰것부터 작은 순으로 정렬한 새로운 정수를 리턴해주세요.
# n은 1이상 8000000000 이하인 자연수

# 1. 숫자 n의 각 자리 숫자를 문자열 n으로 바꾸어 리스트로 만든다.
```

```python
def solution(n):
    x = list(str(n))  # 문자열로 만든 n을 리스트로 만들고 변수 x에 저장: ["1","1","8","3","7","2"]
    x.sort(reverse=True)  # sort()함수는 리스트(x) 역순 정렬: 크기순
    return int("".join(x))  # 리스트(x)를 다시 합치고 정수로 형변환해서 반환한다.


n = 118372
print(solution(n))
```

```python
def solution(n):
    return int("".join(sorted(str(n), reverse=True)))


n = 118372
print(solution(n))
```

```python
def solution(n):
    n = sorted([x for x in str(n)], key=lambda x: int(x), reverse=True)  # 람다 함수 사용
    n = int("".join(n))
    return n


print(solution(118372))
```

```python
# 21
# 하샤드 수
# 양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 한다.
# x를 입력받아 x가 하샤드 수인지 검사하는 함수
# x 는 1이상, 10000 이하인 정수
def solution(x):
    # 1. 자릿수의 합 구하기
    # 숫자를 문자로 변환('18') -> 한 글자씩 떼어내서 숫자로 바꾼 뒤 더함
    digit_sum = 0
    for i in str(x):
        digit_sum += int(i)
    
    # 2. 나누어 떨어지는지 확인 (나머지가 0인지)
    if x % digit_sum == 0:
        return True
    else:
        return False

# 테스트
print(solution(10))  # True (1+0=1, 10은 1로 나누어떨어짐)
print(solution(11))  # False (1+1=2, 11은 2로 안 나누어떨어짐)
print(solution(18))  # True (1+8=9, 18은 9로 나누어떨어짐)
```
