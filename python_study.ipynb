```python
nums = [10, 20, 30, 40, 50]
# 마지막 값을 출력하세요
print(nums[-1])
# 마지막 값을 999로 바꾸고, 바뀐 리스트 전체를 출력하세요
nums[-1] = 999
print(nums)
```

```python
# 튜플
t = (1, 2, 3, 4, 5)
print(len(t))  # 튜플의 길이 구하기
print(t[1:3])
```

```python
student = {"name": "Kim", "age": 20, "major": "CS"}
print(student["name"])
student["age"] = 2
print(student)
```

```python
a = [1, 2, 3]
b = [4, 5]
# a에 b의 모든 원소를 이어붙인 새로운 리스트 c를 만들고 출력하되, a자체는 변경되지 않도록 만드세요
c = (a) + b
print(c)
print(a)
```

```python
nums = [3, 10, 7, 8, 2, 9, 12]
result = []
for n in nums:
    if n % 2 == 0:
        result.append(n)
print(result)
```

```python
a = [1, 2, 3]
b = a
b.append(4)
# a와 b가 똑같은 리스트 하나를 공유하고 있음
# a가 가라키는 방을 b도 가리키고 있기 때문
print(a)
print(b)
# 따라서 a리스트와 b리스트 모두 [1,2,3,4]가 출력됨
```

```python
a = [1, 2, 3]
b = a[:]
b.append(4)
# a를 슬라이싱해서 b에 복사
print(a)
print(b)
```

```python
t = (1, 2, 3)
t = t + (4,)
print(t)
```

```python
d = {"a": 1, "b": 2}
print(d.get("c"))  # 키 "c"가 없으면 기본값 None 반환
print(d.get("a", 0))  # 키 "a"가 있으므로 값 1 반환
```
